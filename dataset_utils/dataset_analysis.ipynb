```python
import json

with open("instances_default.json" ,'r') as file:
    a = json.load(file)
```

```python
images = a['images']
annotations= a['annotations']
images_length = len(images)
image_ids = [x['image_id'] for x in a['annotations']]
images_with_objects = set(image_ids)
images_with_objects_length = len(images_with_objects)
```

```python
import matplotlib.pyplot as plt

labels = f'Images with Screws {images_with_objects_length}', f'Images without Screws : {images_length - images_with_objects_length}'
sizes = [images_with_objects_length , images_length - images_with_objects_length]
explode = (0.05 , 0)
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=True, startangle=90)
```

```python
annotaions_length = len(annotations)
print(annotaions_length)
average_instances = annotaions_length/len(images)
average_instances
```

```python
bbox = [x['bbox'] for x in a['annotations']]
object_height = [x[3] for x in bbox]
sorted(object_height , reverse=True)
counts, edges, bars = plt.hist(object_height , label='Object height')
plt.bar_label(bars)

plt.show() 
```

```python
relaive_height = [x/3008 for x in object_height] 
sorted(relaive_height)
counts, edges, bars = plt.hist(relaive_height , label='Object height')
plt.bar_label(bars)

plt.show() 
```
